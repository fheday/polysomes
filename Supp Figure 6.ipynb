```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from polan import compare_profiles, fp2poly, plot_poly
```

<h1>Comparison of <i>ncs2</i> deletion mutants from different studies</h1>
<p>Relevant datasets include GSE45366 (cluster 3), GSE67387 (cluster 2), GSE100626 (cluster 1) and GSE124428 (cluster 4).</p>

```python
#read in the datasets
GSE45366 = pd.read_csv('Data/GSE45366_all_data.csv')
GSE67387 = pd.read_csv('Data/GSE67387_all_data.csv')
GSE100626 = pd.read_csv('Data/GSE100626_all_data.csv')
GSE124428 = pd.read_csv('Data/GSE124428_all_data.csv')
```

```python
fig,ax = compare_profiles(dats1 = GSE45366,dats2 = GSE45366,dats1_columns = ['ORF','WT_RNA','WT_Ribo'],dats2_columns = ['ORF','ncs2_RNA','ncs2_Ribo'],conditions = ['WT',r'$\Delta$ncs2'])
fig.suptitle('GSE45366, cluster 3')
#plt.savefig('ncs2_GS45366.pdf')
plt.show()
```

```python
fig,ax = compare_profiles(dats1 = GSE67387,dats2 = GSE67387,dats1_columns = ['ORF','WT_RNA','WT_Ribo'],dats2_columns = ['ORF','ncs2_RNA','ncs2_Ribo'],conditions = ['WT',r'$\Delta$ncs2'])
fig.suptitle('GSE67387, cluster 2')
#plt.savefig('ncs2_GSE67387.pdf')
plt.show()
```

```python
fig,ax = compare_profiles(dats1 = GSE100626,dats2 = GSE100626,dats1_columns = ['Transcript','WT_RNA','WT_Ribo'],dats2_columns = ['Transcript','ncs2_RNA','ncs2_Ribo'],conditions = ['WT',r'$\Delta$ncs2'])
fig.suptitle('GSE100626 cluster 1')
#plt.savefig('ncs2_GSE100626.pdf')
plt.show()
```

```python
fig,ax = compare_profiles(dats1 = GSE124428,dats2 = GSE124428,dats1_columns = ['ORF','WT_RNA','WT_Ribo'],dats2_columns = ['ORF','ncs2_RNA','ncs2_Ribo'],conditions = ['WT',r'$\Delta$ncs2'])
fig.suptitle('GSE124428, cluster 4')
#plt.savefig('ncs2_GS124428.pdf')
plt.show()
```
